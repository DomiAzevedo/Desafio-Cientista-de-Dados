## Problemas

### 1. Valores (str) que deveriam ser booleanos

Os problemas percebidos logo de cara são a maioria das colunas, que são strings, apesar de que há muitas colunas binárias que deveriam ser consideradas booleanos, pois indicam verdadeiro ou falso, sendo elas `obito`, `luz_eletrica`, `em_situacao_de_rua`, `possui_plano_saude`, `familia_beneficiaria_auxilio_brasil`, elas tem os valores `0`, `1`, `TRUE`, `FALSE`, então deve ser padronizado e transformado em booleano.

```python
print(df['obito'].unique())
print(df['luz_eletrica'].unique())
print(df['em_situacao_de_rua'].unique())
print(df['possui_plano_saude'].unique())
print(df['familia_beneficiaria_auxilio_brasil'].unique())
```

### 2. Valores (float) que deveriam ser booleanos
Também há a coluna frequenta_escola, que está como inteira mas os valores são apenas 0 ou 1, então deveria entrar como booleano para não confundir com valores inteiros.



```python
print(df['frequenta_escola'].unique())
print(df['frequenta_escola'].dtype)
```

### 3. Valores que parecem ser booleanos mas são categóricos
Temos um outro problema ainda com uma dessas colunas que é a coluna crianca_matriculada_creche_pre_escola, não é possível saber se 1/TRUE ou 0/FALSE é pré-escola ou creche, deveria ser categórico assim como as colunas sexo e tipo.

```python
print(df['crianca_matriculada_creche_pre_escola'].unique())
```

```python
print(df['tipo'].unique())
print(df['sexo'].unique())
```

### 4. Datas
Ainda temos um problema relacionado às datas, elas estão como string, e queremos que elas sejam identificadas como datas, então isso também deve ser modificado nas colunas `data_cadastro`, `data_nascimento`, `data_atualizacao_cadastro` e `updated_at`.

```python
df[['data_cadastro', 'data_nascimento', 'data_atualizacao_cadastro', 'updated_at']].dtypes
```

### 5. Listas que não estão no formato certo
Esse problema é referente às colunas meios_transporte, doencas_condicoes, meios_comunicacao e em_caso_doenca_procura, que são listas mas estão como strings pois algumas estão dentro de '[]', mas outras não, então deveriam todas serem padronizadas como listas de strings.

```python
df[['meios_transporte', 'doencas_condicoes', 'meios_comunicacao', 'em_caso_doenca_procura']].head(10)
```

### 6. Identidade de gênero
Além disso, aparentemente a coluna `identidade_genero` foi preenchida de maneira errônea, contendo valores como `Sim`, `Não` e valores referentes à orientação sexual como `Heterossexual` e `Homossexual`, esses valores não são conclusivos e não são o que a coluna pede, então devem ser trocados por algo como `Não informado` (assim como os valores nulos dessas colunas). Há também valores conclusivos mas que poderiam ser melhor explicados como `Cis` que poderia diferenciar entre mulher cis e homem cis, e `Travesti` que é um termo ofensivo para se referir à pessoas trans, então esses valores devem ser mudados com base na coluna `sexo`.

```python
print(df['identidade_genero'].unique())
```

```python
print(df['orientacao_sexual'].unique())
```

### 7. Float que deveriam ser int
Há dados em que não podem ser floats, eles sempre serão inteiros, como `pressao_sistolica`, `pressao_diastolica`, `n_atendimentos_atencao_primaria` e `n_atendimentos_hospital`, então devemos mudar o tipo destes dados para `int`.

```python
df[['pressao_sistolica', 'pressao_diastolica', 'n_atendimentos_atencao_primaria', 'n_atendimentos_hospital']].dtypes
```

### 8. Textos em inglês
Acredito que para um melhor entendimento dos dados, eles precisam estar no mesmo idioma, então seria mais interessante mudar os valores e colunas que estão em inglês para o português, que é o idioma original da tabela. Então a coluna `updated_at` poderia se chamar `atualizado_em`, e os valores da coluna `sexo`, poderiam passar de `male`/`female` a `masculino`/`feminino`.

### 9. Valores nulos que não deveriam ser nulos
O último problema são os valores nulos nas colunas `altura`, `peso`, `pressao_sistolica`, `pressao_diastolica`. Que não é possível inferir, mas deveria ser obrigatório a informação desses dados pelo paciente, pois eles são importantes para um diagnóstico médico. Neste caso não é possível padronizar esses valores nulos, pois são valores numéricos e pessoais, seria possível inferir com base nos dados e tentar prever esses valores baseando-se em pessoas com características semelhantes, mas este não é o objetivo aqui.

### 10. Vários encodings na mesma base de dados

Podemos ver nas colunas `meios_transporte`, `doencas_condicoes` e `em_caso_doenca_procura` temos outro encoding e deve ser trocado para `utf-8` para a padronização.

```python
df[['meios_transporte','doencas_condicoes', 'meios_comunicacao']].head()
```

### 11. Letras maiúsculas acentuadas
Esses caracteres deram problema na cópia dos dados no arquivo csv para o banco de dados, como nas linhas 805 e 1856 e 1938 (considerando os nomes das colunas), na coluna `bairro`, então a solução foi deixar todas as strings com letras minúsculas, antes de passar para o PostgreSQL.

```python
df.iloc[[803, 1854, 1936], 3]
```

```python
# Função para converter todas as letras em minúsculas
def lower_case(text):
    if isinstance(text, str):
        return text.lower()
    return text

df = df.applymap(lower_case)

df.head()
```

```python
df.to_csv("dados.csv", index=False)
```
