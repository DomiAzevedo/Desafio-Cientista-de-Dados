### 7. Vários encodings na mesma base de dados

Podemos ver nas colunas `meios_transporte`, `doencas_condicoes`, `meios_comunicacao` e `em_caso_doenca_procura` temos outro encoding que é o `windows-1252` e deve ser trocado para `utf-8` para a padronização.

```python
df[['meios_transporte','doencas_condicoes', 'meios_comunicacao', 'em_caso_doenca_procura']].head()
```

```python
# Função para encontrar caracteres no formato \uXXXX
def encontrar_caracteres_errados(string):
    return re.findall(r'\\u[0-9a-fA-F]{4}', string)
```

```python
# Um dicionário para armazenar os caracteres codificados e uma string de exemplo para cada coluna
resultados_mt = {}
resultados_dc = {}
resultados_mc = {}
resultados_dp = {}

for index, row in df.iterrows():
    for erro in encontrar_caracteres_errados(row['meios_transporte']):
        if erro not in resultados_mt:
            resultados_mt[erro] = row['meios_transporte']

    for erro in encontrar_caracteres_errados(row['doencas_condicoes']):
        if erro not in resultados_dc:
            resultados_dc[erro] = row['doencas_condicoes']

    for erro in encontrar_caracteres_errados(row['meios_comunicacao']):
        if erro not in resultados_mc:
            resultados_mc[erro] = row['meios_comunicacao']

    for erro in encontrar_caracteres_errados(row['em_caso_doenca_procura']):
        if erro not in resultados_dp:
            resultados_dp[erro] = row['em_caso_doenca_procura']
```

```python
# Vendo quais caracteres especiais existem em cada coluna
print('Coluna - \'meios_transporte\'')
for erro, string in resultados_mt.items():
    print(f"Caractere: {erro} - String: {string}")
print('\n')

print('Coluna - \'doencas_condicoes\'')
for erro, string in resultados_dc.items():
    print(f"Caractere: {erro} - String: {string}")
print('\n')

print('Coluna - \'meios_comunicacao\'')
for erro, string in resultados_mc.items():
    print(f"Caractere: {erro} - String: {string}")
print('\n')

print('Coluna - \'em_caso_doenca_procura\'')
for erro, string in resultados_dp.items():
    print(f"Caractere: {erro} - String: {string}")
```
